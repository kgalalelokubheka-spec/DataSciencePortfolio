# Country Data Web Scraping

This project scrapes country data from [Scrape This Site](https://www.scrapethissite.com/pages/simple/) using Python.

## What it does
- Extracts country name, capital, population, and area
- Stores the data in CSV (`countries.csv`) and JSON (`countries.json`)

## Tools Used
- Python
- requests
- BeautifulSoup
- pandas

## How to Run
1. Ensure Python 3.x is installed
2. Install dependencies: `pip install requests beautifulsoup4 pandas`
3. Run the script: `python scrape_countries.py`
4. Generated files: `countries.csv` and `countries.json`

## Notes
- The website is static, so no need for Selenium
- Pagination is not required for this page


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Get the webpage
url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)

# Step 2: Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Step 3: Find all country blocks
countries = soup.find_all("div", class_="country")

data = []

# Step 4: Loop through each country and extract data
for country in countries:
    name = country.find("h3", class_="country-name").get_text(strip=True)
    capital = country.find("span", class_="country-capital").get_text(strip=True)
    population = country.find("span", class_="country-population").get_text(strip=True)
    area = country.find("span", class_="country-area").get_text(strip=True)

    data.append({
        "Country": name,
        "Capital": capital,
        "Population": population,
        "Area": area
    })

# Step 5: Convert to DataFrame
df = pd.DataFrame(data)

# Step 6: Save to CSV
df.to_csv("countries.csv", index=False)

# Step 7: Save to JSON
df.to_json("countries.json", orient="records", indent=4)

# Step 8: Print first few rows to check
print(df.head())
```
